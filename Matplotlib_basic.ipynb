```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt

#Jupyter magic command
# %matplotlib inline
```

```python
x = np.arange(0,20,0.1)
sin = np.sin(x)
cos = np.cos(x)
```

```python
print(x)
print(sin)
```

```python
# By default, pyplot.plot() draws a line chart
plt.plot(x, sin)
plt.plot(x, cos)

plt.title("Sin(x) VS. Cos(x)")
plt.xlabel('x-value')
plt.ylabel('y-value')
```

```python
# Get Current figure
fig = plt.gcf()

# Get current axes
axes = plt.gca()

# plt.plot(x, sin)
# plt.plot(x, cos)
sin_plt = axes.plot( x, sin )
cos_plt = axes.plot(x,cos)

# plt.title("Sin(x) VS. Cos(x)")
axes.set_title("Sin(x) VS. Cos(x)", color='blue')

# plt.xlabel('x-value')
axes.set_xlabel('x-value', color='blue')

# plt.ylabel('y-value')
axes.set_ylabel('y-value', color='blue')
```

Working with Grid

```python
# Create figure
fig = plt.figure(figsize=(10,8))

# Convert to grid
ax1 = plt.subplot2grid((2,2),(0,0))
ax2 = plt.subplot2grid((2,2),(0,1))
ax3 = plt.subplot2grid((2,2),(1,0), colspan=2)

# Draw chart on ax1
x1 = np.arange(0,10,0.5)
y1 = 3*x1 + 5
ax1.plot(x1, y1)

# Draw chart on ax2
x2 = np.arange(0,10,0.5)
y2 = 8*(x2**2) - 2*x2 -10
ax2.plot(x2, y2, color='orange')

# Draw chart on ax3
x3 = np.arange(-10,10,0.5)
y3 = -(x3**3) + (x3**2) + x3 -10
ax3.plot(x3, y3, color='blue')

rng = default_rng()
y4 = y3 + rng.integers(low=-200, high=200, endpoint=True, size=40)
ax3.scatter(x3,y4,color='brown')
```
